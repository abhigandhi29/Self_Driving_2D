```python


import matplotlib.pyplot as plt
import numpy as np
gen=list(range(1,16))
```

```python
f = open("output1.txt")

data = []
diction = {}
for line in f.readlines():
  if line.startswith('racetrack'):
      start = line.find(':')
      end = line.find('\\')
      diction['racetrack'] = line[start+1:end]
  if line.startswith('evolution.js:156'):
    space_ind = line.find(' ')
    #print(space_ind)
    bs_ind = line.find('\\')
    #print(bs_ind)
    diction['cf'] = float(line[space_ind+1:bs_ind])

  if line.startswith('evolution.js:157'):
    space_ind = line.find(' ')
    #print(space_ind)
    bs_ind = line.find('\\')
    #print(bs_ind)
    diction['mp'] = float(line[space_ind+1:bs_ind])
    
  if line.startswith('evolution.js:158'):
    start = line.find('[')
    #print(space_ind)
    end = line.find(']')+1
    #print(bs_ind)
    diction['gencounts'] = eval(line[start:end])
    data.append(diction.copy())
```
