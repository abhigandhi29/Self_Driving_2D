```python
f = open("output.txt")

data = []
diction = {}
for line in f.readlines():
  if line.startswith('evolution.js:135'):
    space_ind = line.find(' ')
    #print(space_ind)
    bs_ind = line.find('\\')
    #print(bs_ind)
    diction['cf'] = float(line[space_ind+1:bs_ind])

  if line.startswith('evolution.js:136'):
    space_ind = line.find(' ')
    #print(space_ind)
    bs_ind = line.find('\\')
    #print(bs_ind)
    diction['mp'] = float(line[space_ind+1:bs_ind])
  
  if line.startswith('evolution.js:142'):
    space_ind = line.find(' ')
    #print(space_ind)
    bs_ind = line.find('\\')
    #print(bs_ind)
    diction['malvalavg'] = float(line[space_ind+1:bs_ind])

  if line.startswith('evolution.js:143'):
    start = line.find('[')
    #print(space_ind)
    end = line.find(']')+1
    #print(bs_ind)
    diction['maxvals'] = eval(line[start:end])
    
  if line.startswith('evolution.js:144'):
    start = line.find('[')
    #print(space_ind)
    end = line.find(']')+1
    #print(bs_ind)
    diction['gencounts'] = eval(line[start:end])
    print(diction)
    data.append(diction.copy())
    
    
    
    
```
